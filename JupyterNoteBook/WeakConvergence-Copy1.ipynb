```python
# Import numpy and matplotlib, and use jupyter magic to
# get plots directly in notebook
import numpy as np
import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt
# Get nicer looking plots than default
plt.style.use('bmh')
# Timer to measure the performance of methods
from time import time

#np.seterr(all = 'raise')
```

```python
################
#### Visser ####
################

def ddz(K, z):
    # This function calculates a numerical derivative of K(z),
    # using central finite difference
    dz = 1e-7
    return (K(z+dz/2) - K(z-dz/2))/dz

def d2dz(K, z):
    # This function calculates a numerical second derivative of K(z),
    # using central or one-sided finite difference
    dz = 1e-5
    return (K(z+dz) - 2*K(z) + K(z-dz))/dz**2

def d3dz(K, z):
    # This function calculates a numerical second derivative of K(z),
    # using central or one-sided finite difference
    dz = 1e-3
    return (K(z+2*dz) - 2*K(z+dz) + 2*K(z-dz) - K(z-2*dz))/(2*dz**3)

def milstein2(K, z, w, dt, dW = None):
    if dW is None:
        dW = np.random.normal(loc = 0, scale = np.sqrt(dt), size = z.size)
    dKdz     = ddz( K, z)
    ddKddz   = d2dz(K, z)
    dddKdddz = d3dz(K, z)
    sqrt2K   = np.sqrt(2*K(z))
    a = w + dKdz
    b = sqrt2K
    #  bb'  = K'(z) (because the sqrt cancels)
    bb = dKdz
    # (ab)' = w * K' / sqrt(2K) + K'' * sqrt(K) + (K')**2 / sqrt(2K)
    ab = ddKddz * sqrt2K + (w+dKdz)*dKdz/sqrt2K
    # b''b**2 = K'' * sqrt(2K) - (K')**2 / sqrt(2K)
    bb2 = ddKddz*sqrt2K - dKdz**2/sqrt2K
    # aa' = (w+K')K''
    aa = (w+dKdz)*ddKddz
    # a''b**2 = K''' * 2K
    ab2 = dddKdddz * sqrt2K**2
    return z + a*dt + b*dW + bb*(dW**2 - dt)/2 + (ab + bb2/2)*dW*dt/2 + (aa + ab2/2)*dt**2/2
    
def milstein1(K, z, w, dt, dW = None):
    if dW is None:
        dW = np.random.normal(loc = 0, scale = np.sqrt(dt), size = z.size)
    dKdz = d1Kdz(z)
    # bb' = K'(z) (because the sqrt cancels)
    return z + (w + dKdz)*dt + np.sqrt(2*K(z))*dW + dKdz*(dW**2 - dt)/2

def visser(K, z, w, dt, dW = None):
    if dW is None:
        dW = np.random.normal(loc = 0, scale = np.sqrt(dt), size = z.size)
    dKdz = ddz(K, z)
    return z + (w + dKdz)*dt + np.sqrt(2*K(z+(w + dKdz)*dt/2))*dW

def euler(K, z, w, dt, dW = None):
    if dW is None:
        dW = np.random.normal(loc = 0, scale = np.sqrt(dt), size = z.size)
    dKdz = d1Kdz(z)
    return z + (w + dKdz)*dt + np.sqrt(2*K(z))*dW

def taylor15(K, z, w, dt, dW = None, dZ = None):
    if (dW is None) or (dZ is None):
        # Generate correlated random numbers, using
        # method from Kloeden & Platen, page 352
        U = np.random.normal(loc = 0, scale = 1, size = (2, z.size))
        dW = U[0,:]*np.sqrt(dt)
        dZ = (1/2)*dt**(3/2)*(U[0,:] + U[1,:]/np.sqrt(3))
    
    d1K    = d1Kdz(z)
    d2K    = d2Kdz(z)
    d3K    = d3Kdz(z)
    sqrt2K = np.sqrt(2*K(z))
    a      = w + d1K
    b      = sqrt2K
    # bb'  = K'(z) (because the sqrt cancels)
    bbp    = d1K
    # a'b  = K''*sqrt(2K) (assuming w' = 0)
    apb    = d2K * b
    # aa'  = (w+K')K''
    aap    = (w+d1K)*d2K
    # a''b**2 = K''' * 2K
    appb2  = d3K * sqrt2K**2
    # ab'  = (w + K')K/sqrt(2K)
    abp    = (w + d1K)*d1K/sqrt2K
    # b''b**2 = K'' * sqrt(2K) - (K')**2 / sqrt(2K)
    bppb2  = d2K*sqrt2K - d1K**2/sqrt2K
    # b(b')**2 = bb' * b' = K' * K'/sqrt(2K)
    bbp2   = d1K**2/sqrt2K
    return z + (a*dt + b*dW + bbp*(dW**2-dt)/2 + apb*dZ + (aap+appb2/2)*dt**2/2 
             + (abp + bppb2/2)*(dW*dt-dZ) + (bppb2 + bbp2)*(dW**2/3 - dt)*dW/2 )


###########################
#### Utility functions ####
###########################

def advect(z, v, dt):
    # Move all particles
    # Assuming constant speed (at least within a timestep)
    return z + dt*v

def reflect(z, bottom = None):
    # Reflect from surface and optionally bottom
    # Depth is positive downwards
    # Reflect from surface
    z = np.abs(z)
    if bottom is not None:
        # Reflect from bottom
        z[z > bottom] = 2*bottom - z[z > bottom]
    return z


##############################
#### Diffusivity profiles ####
##############################

def linearexponential(k0, z0, z):
    # Sort-of realistic looking diffusivity profile II
    # Increases linearly from surface, then decays
    # Reaches a maximum of k0 at depth z0
    return k0*(z/z0)*np.exp(-z/z0)/np.exp(-1)
```

```python
Tmax = 6*3600 + 1
dt   = 60
Nt   = int(Tmax/dt)
#### Max depth ####
Zmax = 20
#### number of particles ####
Np   = 1000000
#### Sinking speed ####
w = 0.0
#### Diffusivity profile ####
k0 = 1e-6
k1 = 1e-4
z0 = 10.0
a  = 1.0

# Analytical derivatives, obtained from sympy
#K = lambda z: k0 + np.exp(1)*k1*z*np.exp(-z/z0)/z0
#d1Kdz = lambda z: -np.exp(1)*k1*z*np.exp(-z/z0)/z0**2 + np.exp(1)*k1*np.exp(-z/z0)/z0
#d2Kdz = lambda z: np.exp(1)*k1*(z/z0 - 2)*np.exp(-z/z0)/z0**2
#d3Kdz = lambda z: np.exp(1)*k1*(-z/z0 + 3)*np.exp(-z/z0)/z0**3

K     = lambda z: k0 + (-k0 + k1)/(np.exp(a*(z - z0)) + 1)
d1Kdz = lambda z: -a*(-k0 + k1)*np.exp(a*(z - z0))/(np.exp(a*(z - z0)) + 1)**2
d2Kdz = lambda z: a**2*(1 - 2*np.exp(a*(z - z0))/(np.exp(a*(z - z0)) + 1))*(k0 - k1)*np.exp(a*(z - z0))/(np.exp(a*(z - z0)) + 1)**2
d3Kdz = lambda z: a**3*(k0 - k1)*(1 - 6*np.exp(a*(z - z0))/(np.exp(a*(z - z0)) + 1) + 6*np.exp(2*a*(z - z0))/(np.exp(a*(z - z0)) + 1)**2)*np.exp(a*(z - z0))/(np.exp(a*(z - z0)) + 1)**2

mu = 10
sigma = 0.2
Z0 = np.random.normal(loc = mu, scale = sigma, size = Np)

Z = Z0.copy()
tic = time()
for i in range(Nt):
    Z = visser(K, Z, w, dt)
    Z = reflect(Z, bottom = Zmax)
    Z = advect(Z, w, dt)
toc = time()
print('Time spent: ', toc - tic)
```

```python
Tmax = 12*3600 + 1
#### Max depth ####
Zmax = 20
#### number of particles ####
Np   = 20000000
#### Sinking speed ####
w = 0.0
#### Diffusivity profile ####
k0 = 1e-6
k1 = 1e-4
z0 = 10.0
a  = 1.0

# Analytical derivatives, obtained from sympy
#K = lambda z: k0 + np.exp(1)*k1*z*np.exp(-z/z0)/z0
#d1Kdz = lambda z: -np.exp(1)*k1*z*np.exp(-z/z0)/z0**2 + np.exp(1)*k1*np.exp(-z/z0)/z0
#d2Kdz = lambda z: np.exp(1)*k1*(z/z0 - 2)*np.exp(-z/z0)/z0**2
#d3Kdz = lambda z: np.exp(1)*k1*(-z/z0 + 3)*np.exp(-z/z0)/z0**3

K     = lambda z: k0 + (-k0 + k1)/(np.exp(a*(z - z0)) + 1)
d1Kdz = lambda z: -a*(-k0 + k1)*np.exp(a*(z - z0))/(np.exp(a*(z - z0)) + 1)**2
d2Kdz = lambda z: a**2*(1 - 2*np.exp(a*(z - z0))/(np.exp(a*(z - z0)) + 1))*(k0 - k1)*np.exp(a*(z - z0))/(np.exp(a*(z - z0)) + 1)**2
d3Kdz = lambda z: a**3*(k0 - k1)*(1 - 6*np.exp(a*(z - z0))/(np.exp(a*(z - z0)) + 1) + 6*np.exp(2*a*(z - z0))/(np.exp(a*(z - z0)) + 1)**2)*np.exp(a*(z - z0))/(np.exp(a*(z - z0)) + 1)**2

mu = 10
sigma = 0.2
Z0 = np.random.normal(loc = mu, scale = sigma, size = Np)

timesteps = np.array([600, 900, 1200, 1800, 2700, 3600, 5400, 7200, 10800, 14400, 21600])
    
Zes = []
for dt in timesteps:
    Nt   = int(Tmax/dt)
    Z = Z0.copy()
    tic = time()
    for i in range(Nt):
        Z = euler(K, Z, w, dt)
        Z = reflect(Z, bottom = Zmax)
        Z = advect(Z, w, dt)
    toc = time()
    Zes.append(Z)
    print('Time spent: ', toc - tic)
    
Zms = []
for dt in timesteps:
    Nt   = int(Tmax/dt)
    Z = Z0.copy()
    tic = time()
    for i in range(Nt):
        Z = milstein2(K, Z, w, dt)
        Z = reflect(Z, bottom = Zmax)
        Z = advect(Z, w, dt)
    toc = time()
    Zms.append(Z)
    print('Time spent: ', toc - tic)
```
